```python
import os
import requests
from dotenv import load_dotenv
```

```python
load_dotenv()
API_KEY = os.getenv("KAKAO_API_KEY")
```

```python
def search_places(keyword, page=1, size=15):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json"
    headers = {"Authorization" : f"KakaoAK {API_KEY}"}
    params = {"query": keyword, "page":page, "size":size}
    resp = requests.get(url, headers = headers, params = params)
    resp.raise_for_status()
    return resp.json()
```

```python
import pandas as pd
```

```python
def docs_to_df(documents):
    rows = []
    for d in documents:
        rows.append({
            "id" : d.get("id"),
            "place_name" : d.get("place_name"),
            "x" : float(d.get("x")) if d.get("x") else None,
            "y" : float(d.get("y")) if d.get("y") else None,
        })
    return pd.DataFrame(rows)
```

```python
dfs = []
page =1

while True:
    res = search_places("서경대 맛집", page=page, size =10)
    df = docs_to_df(res["documents"])
    dfs.append(df)
    
    if res["meta"]["is_end"]:
        break
    page+=1
```

```python
result = pd.concat(dfs, ignore_index =True)
```

```python
print(result.shape)
```

```python
result.head()
```

```python
result.to_csv(f"서경대맛집_places.csv", index = False, encoding='utf-8')
```
