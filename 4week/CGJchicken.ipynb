```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from itertools import count
import xml.etree.ElementTree as ET
import ssl
```

```python
def get_request_url(url, enc='utf-8'):
    req = urllib.request.Request(url)
    
    try:
        ssl._create_default_https_context = ssl._create_unverified_context # ssl 추가
        
        response = urllib.request.urlopen(req)
        if response.getcode() == 200: # 성공시
            try:
                rcv = response.read()
                ret = rcv.decode(enc)
            except UnicodeDecodeError:
                ret= rcv.decode(enc, 'replace')
            return ret
    except Exception as e:
        print(e)
        return None
```

```python
def CheogajipAddress(result):
    for page_idx in range(0,125):
        Cheogajip_URL= "https://www.cheogajip.co.kr/bbs/board.php?bo_table=store&page=%s" %str(page_idx+1)
        print(Cheogajip_URL)
        response = urllib.request.urlopen(Cheogajip_URL)
        soupData = BeautifulSoup(response, 'html.parser') # html 파싱
        tbody_tag = soupData.find('tbody') # tbody 태그 가져오기
        
        for store_tr in tbody_tag.findAll('tr'):
            tr_tag = list(store_tr.strings)
            store_name = tr_tag[1]
            store_address = tr_tag[3]
            store_sido_gu = store_address.split()[:2] # 시랑 구 구분짓기
            store_phone = tr_tag[5]
            result.append([store_name] + store_sido_gu + [store_address] + [store_phone])
            
        
```

```python
def cswin_Cheogajip():
    result = []
    col_name = ['store', 'sido', 'gungu', 'store_address', 'store_phone']
    CheogajipAddress(result)
    cheogajip_table = pd.DataFrame(result, columns=col_name)
    cheogajip_table.to_csv('cheogajip.csv', encoding = 'utf-8', index = True)
    del result[:]
    
    print("끝")
```

```python
cswin_Cheogajip()
```
