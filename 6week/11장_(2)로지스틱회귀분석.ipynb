```python
import numpy as np
import pandas as pd
```

```python
feature_name_df = pd.read_csv('UCI_HAR_Dataset/features.txt', sep = '\s+', header = None,names = ['index', 'feature_name'], engine = 'python')
```

```python
# 상위 다섯개만
feature_name_df.head()
```

```python
feature_name_df.shape
```

```python
# feature_name만 가져옴
feature_name = feature_name_df.iloc[:, 1].values.tolist()
```

```python
feature_name
```

```python
feature_name[:5]
```

```python
X_train = pd.read_csv('UCI_HAR_Dataset/train/X_train.txt', delim_whitespace = True, header = None, encoding = 'latin-1')
X_train.columns = feature_name

X_test = pd.read_csv('UCI_HAR_Dataset/test/X_test.txt', delim_whitespace = True, header = None, encoding = 'latin-1')
X_test.columns = feature_name

Y_train = pd.read_csv('UCI_HAR_Dataset/train/y_train.txt', sep = '\s+', header = None, names = ['action'], engine = 'python')

Y_test = pd.read_csv('UCI_HAR_Dataset/test/y_test.txt', sep = '\s+', header = None, names = ['action'], engine = 'python')
```

```python
X_train.head()
```

```python
print(Y_train['action'].value_counts())
```

```python
label_name_df = pd.read_csv('UCI_HAR_Dataset/activity_labels.txt', sep = '\s+', header = None, names = ['index', 'label'], engine = 'python')
```

```python
label_name = label_name_df.iloc[:,1].values.tolist()
```

```python
label_name
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df_HAR = DecisionTreeClassifier(random_state = 156)
```

```python
df_HAR.fit(X_train, Y_train)
```

```python
Y_predict = df_HAR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
print('결정 트리 예측 정확도: {0:.4f}'.format(accuracy))
```

```python
print('결정 트리의 현재 하이퍼 매개변수: \n', df_HAR.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 하이퍼파라미터간 좋은 결과를 낼 수 있는 것을 auto -> 최적의 하이퍼 매개변수를 찾기
params = {
    'max_depth' :[6,8,10,12,16,20,24]
}

grid_cv = GridSearchCV(df_HAR, param_grid = params, scoring = 'accuracy', cv=5, return_train_score = True)

grid_cv.fit(X_train, Y_train)
```

```python
print('최고 평균 정확도: {0:.4f}, 최적 하이퍼 매개변수: {1}'.format(grid_cv.best_score_, grid_cv.best_params_))
```

```python
params = {
    'max_depth' :[8, 16, 20],
    'min_samples_split': [8, 16, 24]
}

grid_cv = GridSearchCV(df_HAR, param_grid = params, scoring = 'accuracy', cv = 5, return_train_score = True)
grid_cv.fit(X_train, Y_train)
```

```python
print('최고 평균 정확도: {0:.4f}, 최적 하이퍼 매개변수 : {1}'.format(grid_cv.best_score_, grid_cv.best_params_))
```

```python
best_dt_HAR = grid_cv.best_estimator_
best_Y_predict = best_dt_HAR.predict(X_test)
best_accuracy = accuracy_score(Y_test, best_Y_predict)

print('best 결정 트리 예측 정확도: {0:.4f}'.format(best_accuracy))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 각 피처의 중요도를 알아내기
feature_importance_values = best_dt_HAR.feature_importances_
feature_importance_values_s = pd.Series(feature_importance_values, index = X_train.columns)
```

```python
feature_top10 = feature_importance_values_s.sort_values(ascending=False)[:10]
```

```python
plt.figure(figsize=(10,5))
plt.title('Feature Top 10')
sns.barplot(x = feature_top10, y = feature_top10.index)
plt.show()
```
